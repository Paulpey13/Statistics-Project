**Exercise 1.** Define a 10-fold classification strategy to test the accuracy of a Linear Discriminant Analysis (LDA) classifier for the data created as follows:

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples = 100, n_features=2, n_redundant=0, n_informative=2, \
                           random_state=0, n_clusters_per_class=1, weights = [0.5])
```

Check what is inside X and y :

```python
print("X :\n",X[0:5])
print("\ny :\n",y[0:5])
```

Define number of folds and LDA :

```python
n_folds = 10
LDA = LinearDiscriminantAnalysis()
```

Define the function `reduce_data()` to compute correlation between features and the target and only selecting the intersting features.

```python
def reduce_data( X, y, features_to_retain):
    N_features = X.shape[1]
    # We compute the correlation between each feature and the target y
    correlation = np.array([np.abs(np.corrcoef(X[:,i],y)[0,1]) for i in range(N_features)])

    index_sorted = np.argsort(correlation)
    # We select only the features with highest correlation
    selected = index_sorted[-features_to_retain:]

    # The output is the data matrix with reduced dimensionality    
    reduced_X = X[:,selected]    
    
    return reduced_X, y, selected
```

Let's perform a handmade cross validation :

First, we use `StratifiedKFold()` with 10 folds, 

Then we loop 50 times to test the pipeline :

    •Generate data
    •We reduce it and perform the 10-fold cross validation :
        •Each time we fit LDA with the train set
        •Then we predict y_test with X_test and compute the accuracy
        
    •At the end of the i loop, we put the previously computed accuracy in the accuracy list

Finally, we can compute the mean accuracy.

```python
cv = StratifiedKFold(n_splits = n_folds)
features_to_retain=100
cv_accuracy = []


# We test the pipeline 50 times
for i in range(50):
    # Data generation
    X_, y_ = make_classification(n_samples = 100, n_features=2, n_redundant=0, n_informative=2, \
                           random_state=0, n_clusters_per_class=1, weights = [0.5])
    # Dimensionality reduction
    X, y, _ = reduce_data(X_, y_, features_to_retain)
    
    accuracy = []
    # 10-fold cross-validation
    for train, test in cv.split(X, y):
        LDA.fit(X[train], y[train])
        accuracy.append(LDA.score(X[test], y[test]))

    # Store the accuracy for iteration i        
    cv_accuracy.append(accuracy)

print("Mean of 10 fold cross-validation score :",round(np.mean(cv_accuracy),2))
```

```python
print(list(set(tuple(lst) for lst in cv_accuracy)))
```

We can also see in the previous cell that all the accuracy are the same and there is on **only one unique value**. 


We can verify our results using `cross_val_score` from `sklearn` to perform 10-fold classification strategy to test the accuracy of a Linear Discriminant Analysis, using the metric accuracy

```python
X, y = make_classification(n_samples = 100, n_features=2, n_redundant=0, n_informative=2, \
                           random_state=0, n_clusters_per_class=1, weights = [0.5])
```

```python
cv_score = cross_val_score(estimator=LDA,X=X,y=y,scoring="accuracy",cv=n_folds)
print("10 fold cross-validation score :",cv_score)
print("Mean of 10 fold cross-validation score :",round(np.mean(cv_score),2))
```

Here, we see that we obtain the same result with the `hand-made cross validation` and the `cross_val_score` function from `sklearn`



__Exercise 2.__ Use the previous 10-fold cross-validation to plot and compute the average area under the curve of the LDA classifier. You can use the built in method *predict_proba(X)*

### First, let's plot the ROC curves for each folds :

Here, as previously, we perform k-fold cross-validation with `StratifiedKFold()` and 10 folds.

We loop over each folds to fit LDA using train data the, we compute the probabilities with `predict_proba()` for the test data.

These probabilites will be used to compute the ROC curve and its associated AUC value for the current fold.

Each curves are ploted and at the end, we compute the mean of all AUCs of the 10 folds.



```python
cv = StratifiedKFold(n_splits=n_folds)
#cv=RepeatedKFold(n_splits=n_folds)
roc_sample = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)
i = 0

for train, test in cv.split(X, y):
    probas_ = lda.fit(X[train], y[train]).predict_proba(X[test])
    # Compute ROC curve and area under the curve
    fpr, tpr, thresholds = roc_curve(y[test], probas_[:, 1])
    roc_sample.append(np.interp(mean_fpr, fpr, tpr))
    # We impose the ROC curve to start from 0
    roc_sample[-1][0] = 0.0
    # Compute the area under the curve
    roc_auc = auc(fpr, tpr)
    aucs.append(roc_auc)
    # Plotting the ROC curve
    plt.plot(fpr, tpr, lw=1, alpha=0.3,
             label='ROC fold %d (AUC = %0.2f)' % (i, roc_auc))
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')

    i += 1

plt.legend()
plt.show()

print(print("Mean of area under the curve is :",round((sum(aucs) / len(aucs)),3)))
```

### We then can summarize the cross-validation results :

Here, we summarize the CV results of the previous code by computing the mean ROC curve and again, its associated mean AUC value between the 10 folds. The standard deviation is also computed in order to have an idea of the different AUCs

In order to plot this summary, we first add a diagonal line of random classification to the plot, which can be used as a reference.

We then compute the ROC curves, The AUCs and their respective means and the ROC's std.

Then, we plot this summary to have a very good overview of the mean ROC and its std.


```python
# Summarizing the cross-validation results

# Adding diagonal line of random classification 
plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',
         label='Chance', alpha=.8)

mean_tpr = np.mean(roc_sample, axis=0)
std_tpr = np.std(roc_sample, axis=0)

# We impose the end point of the average ROC to be equal to 1
mean_tpr[-1] = 1.0

mean_auc = auc(mean_fpr, mean_tpr)
std_auc = np.std(aucs)

plt.plot(mean_fpr, mean_tpr, color='b',
         label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
         lw=2, alpha=.8)

tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
plt.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=.2,
                 label=r'$\pm$ 1 std. dev.')

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

In order to define the **appropriate cross-validation strategy** of the AUC, we need to get deeper in the analysis of the data to understand it better and find the best models and parameters to use.

First, let's generate some data from the given generation function.

```python
X, y = make_classification(n_samples = 1000, n_features=5, n_redundant=0, n_informative=3, \
                           random_state=0, n_clusters_per_class=1, weights = [0.9])
```

We can see that the data is very unbalanced with much more 0s than 1s. 

We have to take this parameter into account for the CV strategy.

But after doing some research, I found that the function `StratifiedKFold()` that we were already using is very good for unbalanced dataset.

sources : https://machinelearningmastery.com/cross-validation-for-imbalanced-classification/#:~:text=For%20example%2C%20we%20can%20use,in%20the%20complete%20training%20dataset.


Now that wo solved the unbalanced problem, we need to find the optimal number of folds $k$

```python
means = [] 
stdevs = []
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
folds = list(range(2, 51))

for k in folds:
    model = LinearDiscriminantAnalysis()
    
    accuracies_within_fold = []
    
    skf = StratifiedKFold(n_splits=k)
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]


        model.fit(X_train,y_train)
        accuracies_within_fold.append(accuracy_score(y_test, model.predict(X_test)))
        
    #Average and std for a particular k fold
    means.append(np.mean(accuracies_within_fold))
    stdevs.append(np.sqrt(np.var(accuracies_within_fold, ddof = 1)))

means  = np.asarray(means)
stdevs  = np.asarray(stdevs)

data = {'Fold': folds, 'Mean': means, 'Std Dev': stdevs}

df = pd.DataFrame(data)

#transpose to have a better overview
df.T
```

In this dataframe, we can see that the best $k$ is 13 with the max mean and 3 with the lowers standard deviation :

```python
print("Fold with maximum mean:", df['Fold'].loc[df['Mean'].idxmax()])
print("Highest mean:", df['Mean'].max())

print("Fold with Lowest standard deviation:", df['Fold'].loc[df['Std Dev'].idxmin()])
print("Lowest standard deviation:", df['Std Dev'].min())
```

## Short summary on the [Dangers of Cross-Validation](http://people.csail.mit.edu/romer/papers/CrossVal_SDM08.pdf).
Written by *R. Bharat Rao*, *Glenn Fung* and *Romer Rosales*



The primary objective of a classifier is to learn generalized features from a dataset and make accurate predictions on unseen data. However, obtaining test data can often be problematic, and thus, resampling methods on training data were developed to increase model robustness. One of the popular methods is **cross-validation** (CV), which trains a model on a subset of the available dataset and validates it on the rest. Despite being considered a reliable model comparator, CV can significantly overfit and fail at model comparison. The paper aims to provide a comprehensive demonstration on the dangers of Cross Validation.

The paper employs **Leave-One-Out Cross Validation** (LOOCV) to train and predict M algorithms on a synthetic dataset with a **&frac12;** true error. The accuracy score, an average of three LOOCV iterations, is used to evaluate the performance. The results indicate that accuracy increases as M increases, diverging from the true error **&frac12;**. The paper notes that **overfitting** behavior is also apparent when increasing the sample size, data dimensionality, model complexity, and the number of selected features. LOOCV, in particular, tends to select algorithms that significantly overestimate the true accuracy.

The paper demonstrates and warns that CV can become ineffective, leading to a higher variance, overfitting, and unreliability of the model's generalization error. To address this issue, the paper recommends using a sequestered test set to be used once training and validation are completed, with the error on the validation and test sets checked to ensure their proximity. The paper cautions against using the test set for selection, as it violates the hypothesis of testing in a real-world setting.


**Conclusion**

The paper highlights the limitations of cross-validation as a reliable model comparator, noting its potential for overfitting and suggesting the use of a sequestered test set to improve model robustness.
